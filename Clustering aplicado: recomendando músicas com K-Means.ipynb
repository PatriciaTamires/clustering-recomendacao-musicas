```python
df_gerais = spark.read.option("header", True).csv(
    "/Volumes/workspace/default/raw_musicas/Dados_totais.csv"
)

display(df_gerais)




df_genero = spark.read.option("header", True).csv(
    "/Volumes/workspace/default/raw_musicas/data_by_genres.csv"
)

display(df_genero)



df_anos = spark.read.option("header", True).csv(
    "/Volumes/workspace/default/raw_musicas/data_by_year.csv"
)

display(df_anos)
```

```python
### Contagem de informações dentro do database
df_gerais.count()
df_genero.count()
df_anos.count()
```

```python
### Validacao [Null] nas bases 
### 1 = tem nulo, 0 = não tem nulo

from pyspark.sql.functions import col, when, sum, lit

resultado = []

for c in df_anos.columns:
    qtd_nulos = df_anos.filter(col(c).isNull()).count()
    flag_nulo = 1 if qtd_nulos > 0 else 0

    resultado.append((c, flag_nulo))

df_nulos = spark.createDataFrame(
    resultado,
    ["coluna", "tem_nulo"]
)

display(df_nulos)
```
